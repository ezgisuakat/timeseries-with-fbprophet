```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('municipality_bus_utilization.csv')
print(df)
```

```python
df.info()
```

```python
df.isnull().sum() 
```

```python
df['usage'] = df['usage'].interpolate()  # it is not neccessary because missing data isn't founded.
df['usage']
```

```python
START = 0
STOP = 10
list_usage = []


while True:
    
    if STOP > len(df['usage']):
        break
        
    
    list_usage.append(df['usage'][START:STOP].max())
    START += 10
    STOP += 10
    
    
    
```

```python
print(len(list_usage))
list_usage
```

```python
list_dates = []
COUNT =  0

while True:
    
    if COUNT >= len(df['timestamp']):
        break
        
    
    list_dates.append(df['timestamp'][COUNT])
    COUNT += 10

    
    
```

```python
print(len(list_dates))
list_dates
```

```python
data = {'time' : list_dates, 'number_of_usage' : list_usage}
new_df = pd.DataFrame(data)
```

```python
new_df
```

```python
new_df.info()
```

```python
new_df['time'] = pd.to_datetime(new_df['time'])
```

```python
new_df.info()
```

```python
new_df = new_df.rename(columns = {'time': 'ds', 'number_of_usage': 'y'})
new_df[-268:]
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(new_df['ds'], new_df['y'],'r')
plt.xlabel('ds')
plt.ylabel('y')
plt.show()
```

```python
test_df = new_df[-268:]
test_df
```

```python
train_df = new_df[:-268]
train_df
```

```python
from fbprophet import Prophet
model = Prophet()
model.fit(train_df)
```

```python
prediction = model.predict(test_df)
prediction[['yhat', 'yhat_lower', 'yhat_upper']]
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(test_df['y'], prediction['yhat'])
print(f'mean square error : {mae}')
```

```python
model.plot(prediction)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.plot(test_df['y'].values, label='Actual')
plt.plot(prediction['yhat'].values, label='Predicted')
plt.legend()
plt.show()
```
